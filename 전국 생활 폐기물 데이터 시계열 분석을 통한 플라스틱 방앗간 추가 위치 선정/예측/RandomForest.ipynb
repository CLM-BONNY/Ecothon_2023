# Import packages

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

# Import data

```python
life_all = pd.read_csv('life_all.csv')
# 2020년, 2021년의 단위 맞춰주기
life_all[['2020년','2021년']] = round(life_all[['2020년','2021년']]/365, 1)
life_all=life_all.set_index('구분1')
life_all
```

### Split Train, Test Data

* 1996~2016까지의 추이를 바탕으로 = Training Data
* 2016~2021년도의 상승/하락 예측 = Test Data

```python
train_data, train_label = seoul['서울'][1:-6], seoul['return'][1:-6] # 1997~2016
test_data, test_label = seoul['서울'][-6:], seoul['return'][-6:] # 2016~2021

X_train= pd.DataFrame(train_data)
y_train= pd.DataFrame(train_label)
X_test= pd.DataFrame(test_data)
y_test= pd.DataFrame(test_label)
```

### GridSearchCV

```python
from sklearn.model_selection import cross_val_score, GridSearchCV

forest_parameters = [{"n_estimators" : [2,4,6,8,10],

                      "criterion" : ["gini","entropy"],

                      "min_samples_leaf":[2,4,6,8,10],

                      "random_state":[1]}]
                      
                      
forest_grid = GridSearchCV(estimator = RandomForestClassifier(), param_grid = forest_parameters, scoring ="accuracy", cv=10, n_jobs=1)
forest_grid_result =forest_grid.fit(X_train, y_train)
best_forest_parameters = forest_grid_result.best_params_
forest_score = forest_grid_result.best_score_

print(forest_grid_result.best_params_)
```

```python
def ran_for(x):    
    train_data, train_label = x.iloc[:,0][1:-6], x.iloc[:,1][1:-6] # 1997~2016
    test_data, test_label = x.iloc[:,0][-6:], x.iloc[:,1][-6:] # 2016~2021
    X_train= pd.DataFrame(train_data)
    y_train= pd.DataFrame(train_label)
    X_test= pd.DataFrame(test_data)
    y_test= pd.DataFrame(test_label)

    # GridSearchCV parameter settings
    forest_parameters = [{"n_estimators" : [2,4,6,8,10],"criterion" : ["gini","entropy"],
                          "min_samples_leaf":[2,4,6,8,10],"random_state":[1]}]
    forest_grid = GridSearchCV(estimator = RandomForestClassifier(), param_grid = forest_parameters, 
                               scoring ="accuracy", cv=10, n_jobs=1)
    forest_grid_result =forest_grid.fit(X_train, y_train)
    best_forest_parameters = forest_grid_result.best_params_
    forest_score = forest_grid_result.best_score_
    print(forest_grid_result.best_params_)

    # GridSearchCV with the best hyperparameter
    Forest_model = RandomForestClassifier(n_estimators =best_forest_parameters["n_estimators"],
                                         criterion = best_forest_parameters["criterion"],
                                         min_samples_leaf =best_forest_parameters["min_samples_leaf"],
                                         random_state =1).fit(X_train, y_train)
    y_pred_best = Forest_model.predict(X_test)
    print("accuracy: {}".format(metrics.accuracy_score(y_pred_best, y_test)))
    
for i in [gangwon,gyeonggi,gyeongnam,gyeongbuk,gwangju,daegu,daejeon,busan,seoul,sejong,
          woolsan,incheon,jeonnam,jeonbuk,jeju,chungnam,chungbuk]:
    print(i.columns[0])
    ran_for(i)
```
